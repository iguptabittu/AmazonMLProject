```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt

# Sample dataset loading (replace with your own dataset)
# You can use any publicly available dataset, for example, the 20 Newsgroups dataset
from sklearn.datasets import fetch_20newsgroups

# Load dataset and preprocess it
data = fetch_20newsgroups(subset='all')
X = data.data
y = data.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Naive Bayes": MultinomialNB(),
    "Neural Network (MLP)": MLPClassifier(max_iter=1000)
}

# Dictionary to store results
results = {}

# Train and evaluate models
for model_name, model in models.items():
    # Train the model
    model.fit(X_train_tfidf, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test_tfidf)

    # Store performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred, output_dict=True)
    results[model_name] = {
        "Accuracy": accuracy,
        "Precision (Macro)": class_report['macro avg']['precision'],
        "Recall (Macro)": class_report['macro avg']['recall'],
        "F1-Score (Macro)": class_report['macro avg']['f1-score']
    }

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results).T
results_df = results_df.round(4)  # Round to 4 decimal places for clarity

# Display the results in tabular form
results_df
```

```python
# Import necessary libraries for ensemble
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups

# Load dataset (20 Newsgroups dataset for text classification)
data = fetch_20newsgroups(subset='all')
X = data.data
y = data.target

# Check the dataset
print("Sample Data:", X[:2])
print("Sample Labels:", y[:2])

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize TF-IDF Vectorizer for text feature extraction
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

# Fit and transform the training data
X_train_tfidf = vectorizer.fit_transform(X_train)

# Transform the test data
X_test_tfidf = vectorizer.transform(X_test)

# Initialize individual models: Logistic Regression, Naive Bayes, and Neural Network (MLP)
logreg = LogisticRegression(max_iter=1000, random_state=42)
nb = MultinomialNB()
mlp = MLPClassifier(max_iter=1000, random_state=42)

# Create an ensemble model using Voting Classifier
ensemble_model = VotingClassifier(estimators=[
    ('Logistic Regression', logreg),
    ('Naive Bayes', nb),
    ('Neural Network (MLP)', mlp)
], voting='hard')

# Train the ensemble model
ensemble_model.fit(X_train_tfidf, y_train)

# Predict on the test set using the ensemble model
y_pred = ensemble_model.predict(X_test_tfidf)

# Evaluate the ensemble model
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred, output_dict=True)

# Store the evaluation metrics for the ensemble model
ensemble_results = {
    "Accuracy": accuracy,
    "Precision (Macro)": class_report['macro avg']['precision'],
    "Recall (Macro)": class_report['macro avg']['recall'],
    "F1-Score (Macro)": class_report['macro avg']['f1-score']
}

# Display the results for the ensemble model
print("\nEnsemble Model Evaluation:")
print(ensemble_results)

# Confusion Matrix and ROC Curve for the Ensemble Model
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.colorbar()
    tick_marks = np.arange(len(np.unique(y_true)))
    plt.xticks(tick_marks, tick_marks)
    plt.yticks(tick_marks, tick_marks)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def plot_roc_curve(y_true, y_pred, model_name):
    # Compute ROC curve and ROC area for the Ensemble Model
    fpr, tpr, _ = roc_curve(y_true, y_pred, pos_label=1)
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic Curve for {model_name}')
    plt.legend(loc='lower right')
    plt.show()

# Example: Confusion Matrix and ROC Curve for the Ensemble Model
plot_confusion_matrix(y_test, y_pred, "Ensemble Model")
plot_roc_curve(y_test, y_pred, "Ensemble Model")
```

```python
import matplotlib.pyplot as plt

def plot_training_history(history_rnn, history_lstm):
    # Plot training & validation loss values
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history_rnn.history['loss'], label='RNN Training Loss')
    plt.plot(history_rnn.history['val_loss'], label='RNN Validation Loss')
    plt.plot(history_lstm.history['loss'], label='LSTM Training Loss')
    plt.plot(history_lstm.history['val_loss'], label='LSTM Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 2)
    plt.plot(history_rnn.history['accuracy'], label='RNN Training Accuracy')
    plt.plot(history_rnn.history['val_accuracy'], label='RNN Validation Accuracy')
    plt.plot(history_lstm.history['accuracy'], label='LSTM Training Accuracy')
    plt.plot(history_lstm.history['val_accuracy'], label='LSTM Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')

    plt.tight_layout()
    plt.show()

# Plot the training history
plot_training_history(history_rnn, history_lstm)
```

```python
import nltk
from nltk.util import ngrams
from collections import Counter

corpus=['I love natural language processing','I enjoy programming in python','Natural language processing is fascinating','I love programming and NLP]']
tokenized_corpus=[sentence.split() for sentence in corpus]
trigrams=[list(ngrams(sentence,3,pad_left=True,pad_right=True)) for sentence in tokenized_corpus]
flat_trigrams=[trigram for sentence in trigrams for trigram in sentence]
tri_counts=Counter(flat_trigrams)
def autocomplete(prev_words):
  predictions=[]
  for trigram in tri_counts:
    if trigram[0]==prev_words[0] and trigram[1]==prev_words[1]:
      predictions.append((trigram[2],tri_counts[trigram]))
  predictions=sorted(predictions,key=lambda x:x[1])
  return [word for word,count in predictions[:5]]

print(autocomplete(['I','love']))
```
